```python
import sqlite3
```

```python
#Connection with sqlite database
conn = sqlite3.connect('sales.db')
```

```python
#Interact python command with sqlite database
cursor = conn.cursor()
```

## Create Tables

### Create table customer

```python
try:
    cursor.execute('''
    Create table  Customer (
         customer_id integer primary key autoincrement,
         customer_name text not null,
         customer_address text not null,
         customer_phone text not null,
         customer_email text
     );
     ''')
    print('Customer table created.')
except:
    print('Error:Customer table already exists.')
```

## Create table staff

```python
try:
    cursor.execute('''
        Create table Staff(
            staff_id integer primary key autoincrement,
            staff_name text not null,
            staff_position text not null,
            staff_job_type text not null,
            years_of_experience integer not null
         );
    ''')
    print('Staff table created.')
except:
    print('Staff table already created.')
    
```

## Create table Order

```python
try:
    cursor.execute('''
        Create table Orders (
            order_id integer primary key autoincrement,
            product_name text not null,
            quality integer not null,
            unit_price integer not null,
            order_destination text not null,
            order_date date not null,
            customer_id integer,
            staff_id integer,
            foreign key (customer_id) references Customer(customer_id),
            foreign key (staff_id) references Staff(staff_id)
        );
    ''')
    print('Orders table created.')
except:
    print('Orders table already created.')
```
