```python
try:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))

    num3 = num1 / num2
    print('The addition of the first number and the second number is:', num3)
except ValueError:
    print('Value Error: Only integer numbers are accepted in input field.')
except ZeroDivisionError:
    print('Zero Division Error: Number cannot be divisible by zero.')
```

```python
try:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))

    num3 = num1 / num2
    print('The addition of the first number and the second number is:', num3)
except ValueError:
    print('Value Error: Only integer numbers are accepted in input field.')
except ZeroDivisionError as zde:
    print(zde)           # to print the error that is originally shown by the computer program itself
```

```python
try:
    file_name = input('Enter your file name:')
    with open(f'{file_name}.txt', 'r') as file:
        data = file.read()
        print(data)
except FileNotFoundError:
    print(f' Error: No such file or folder found in directory named as: "{file_name}.txt"')
```

## custom exception

```python
#balance withdrawal
balance = 50000

amount = int(input('Enter your withdrawal amount:'))

if amount < balance:
    balance -= amount
    print(f'Rs. {amount} has been withdrawn from A/C no. 123456789')
else:
    print('Withdrawal amount must be less than balance amount')
```

```python
class BalanceWithdrawlError(Exception):
    pass

class NegativeAmountError(Exception):
    pass
```

```python
def balanceWithdraw():
    balance = 50000
    amount = int(input('Enter your withdrawl amount:'))
    
    if amount > 0:
      if amount < balance:
         balance -= amount
         print(f'Rs, {amount} has been withdrawn from A/C no. 123456789')
      else:
         raise BalanceWithdrawlError(f'Error: Withdrawl amount must be less than balance amount.')
    else: 
        raise NegativeAmountError(f'Error: Amount value must be positive.')
```

```python
try:
    balanceWithdraw()
except BalanceWithdrawlError as e:
    print(e)
except ValueError:
    print('Error: Only integer value are accepted by input field.')
except NegativeAmountError as e:
    print(e)
```
