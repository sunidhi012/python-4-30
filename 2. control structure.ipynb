```python
#tax question
yearly_income= float(input('Enter your yearly income amount:'))

if yearly_income >= 250000 and yearly_income <= 500000:
    tax = yearly_income * 0.05
    print('NOTE: 5% tax is reduced from your yearly income')
elif yearly_income > 500000 and yearly_income <= 1000000:
    tax = yearly_income * 0.1
    print('NOTE: 10% tax is reduced from your yearly income')
elif yearly_income > 1000000:
    tax = yearly_income * 0.2
    print('NOTE: 20% tax is reduced from your yearly income')
elif yearly_income >= 0 and yearly_income < 250000:
    tax = 0
    print('NOTE: No tax is reduced from your yearly income')
else:
    print('Invalid annumal income')
remaining_amount = yearly_income - tax

print(f'Yearly Income: {yearly_income}')
print(f'tax amount: {tax}')
print(f'The remaing amount after deducting the tax is: {remaining_amount}')
print(remaining_amount,'is the remaining amount after deducting the tax ')
    
```

```python
#atm question
pin = 1234
balance = 50000

entered_pin = int(input('Enter the ATM pin'))

if entered_pin == pin:
    print('Access Granted!')
    print('Available Balance:', balance)
    amount = int(input('Enter the withdrawal amount'))
    
    if amount <= 0:
        print('Invalid withdrawal amount')
    elif amount % 1000 != 0:
        print('The amount you are trying to withdraw is not in the multiple of thousands.\nPlease enter the amount in the multiple of thousands.')
    elif amount > balance:
        print('Insufficient Blanace')
    else:
        balance = balance - amount
        print('The withdrawn amount:', amount)
        print('The remaining balance:', balance)
        
else:
    print('Incorrect PIN.ACCESS DENIED')
    
```
