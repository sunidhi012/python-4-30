## Match Case

```python
'''
1. You have ordered Breakfast.
2. You have ordered Lunch.
3. You have ordered Snacks.
4. You have ordered Dinner.
5. You have ordered Dessert.
6. Thank you for choosing us. 
'''
```

```python
choose = int(input('Enter a number: '))

if choose == 1:
    print('You have ordered Breakfast.')
elif choose == 2:
    print('You have ordered Lunch.')
elif choose == 3:
    print('You have ordered Snacks.')
elif choose == 4:
   print('You have ordered Dinner.')
elif choose == 5:
   print('You have ordered Dessert.')
elif choose == 6:
   print('Thank you for choosing us.')
else:
    print('Invalid Number. Please coose order number from 1 to 6.\n Click 6 to exit program.')
```

```python
choose = int(input('Enter a number:')) # case is slower than if

match choose:
    case 1:
        print('You have ordered Breakfast.')
    case 2:
        print('You have ordered Lunch.')
    case 3:
        print('You have ordered Snacks.')
    case 4:
        print('You have ordered Dinner.')
    case 5:
        print('You have ordered Dessert.')
    case 6:
        print('Thank you for choosing us.')
    case _:                 # case _ always comes at last
        print('Invalid Number. Please coose order number from 1 to 6.\n Click 6 to exit program.')
        
```

## break and continue

```python
'''
-Break and continue only can be used inside for or while loop.
- Break -> To stop the program forcefully.
- Continue -> To skip values.
'''
# Print 10 natural numbers but stop program with it finds number 5.
for i in range (1,11):
    if i == 5: # 5 == 5 -> True
        continue
    print(i)
```

## While loop

```python
# until condition becomes false
secret_number = 7
guess_number = 0 #7

while guess_number != secret_number: # the main condition.If this becomes false then the program stops # 7 != 7 -> False
    guess_number = int(input('Enter your guessed number.'))

    if guess_number == secret_number: # 7 == 7 # True
        print(f'Number {guess_number} is guessed correctly.')
    else:
        print('The number you guessed is not correct. Try again.')
```

```python
secret_number = 7
guess_number = 0 #7
remaining_attempt = 3
 
while guess_number != secret_number: 
    print(f'Remaining Attempt: { remaining_attempt}')
    print('-' * 25)
    guess_number = int(input('Enter your guessed number.'))

    if guess_number == secret_number: # 7 == 7 # True
        print(f'Number {guess_number} is guessed correctly.')
    else:
        print('The number you guessed is not correct. Try again.')
        remaining_attempt -= 1 #remaining_attempt = remaining_attempt -1

        if remaining_attempt == 0:
            print(f'{remaining_attempt} attempt left. Time Up!!')
            break
```

```python
pin = 1234
user_pin = 0
balance = 50000
remaining_attempt = 3

while user_pin != pin:
    print(f"Remaining Attempts: {remaining_attempt}")
    user_pin = int(input("Enter your PIN number: "))

    if user_pin == pin:
        print("PIN Verified Successfully")

        while True:
            withdraw_amount = int(input("Enter withdrawal amount: "))

            if withdraw_amount > balance:
                print("Insufficient balance. Try again.")

            elif withdraw_amount % 1000 != 0:
                print("Amount must be in multiples of 1000. Try again.")

            elif withdraw_amount <= 0:
                print("Invalid withdrawal amount. Try again.")

            else:
                balance -= withdraw_amount

                print(f"{withdraw_amount} has been withdrawn from A/C no. 1234567890")
                print(f"Remaining Balance: {balance}")
                break

    else:
        print("Incorrect PIN number!\n")
        remaining_attempt -= 1

    if remaining_attempt == 0:
        print("PIN Blocked. Contact Nearest Branch!!")
        break
```
