## Constructor

```python
'''
__init__()

Types of variable
-----------------------
1. Local Variable -> Variable defined inside constructor
2. Instance Variable -> Variable used to instantiate / store value of local variable.
                     -> Instance variable is also used to pass data from one class to another / one object to another.
3. Global Variable -> Variable defined outside the class or function.
'''
```

```python
SCHOOL_NAME = 'ABC School' # global variable
class Student:
    def __init__(self, student_name, student_address, student_class, student_grade, student_phone): # these are local variable 
     self.student_name = student_name    # to understand, local and instance variable are given the same name, even though it is diff it will not give any error
     self.student_address = student_address 
     self.student_class = student_class 
     self.student_grade = student_grade
     self.student_phone = student_phone

    def student_details(self):
      print(f'School Name: {SCHOOL_NAME}')    # global variable does not need self as it can be accessed by any variable inside or outside the class
      print(f'Name of student: {self.student_name}')
      print(f'Address of student: {self.student_address}') # self. are instance variable
      print(f'Class of student: {self.student_class}')
      print(f'Grade of student: {self.student_grade}')
      print(f'Student Phone Number: {self.student_phone}')
    
```

# Inheritance

```python
'''
- Parent class and Child class
- Child class inherit / extends parent class

Types of Inheritance
----------------------
1. Simple Inheritance(Two Classes)
2. Multiple Inheritance(3 or more than 3 classes)
   -> One parent, Multiple child
3. Multi-level Inheritance(3 or more than 3 classes)
'''
```

```python
'''
- Rules of Inheritance
1. Child Class inherit parent Class.
2. If child class inherit parent class then child class needs to inherit parent class details as well.
'''
```

```python
# customer
# Parent Class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
      self.customer_id = customer_id
      self.customer_name = customer_name
      self.customer_address = customer_address
      self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
# order
#Child Class
class Order(Customer):
    def __init__(self, order_id, order_date, order_price, order_destination, customer_id, customer_name, customer_address, customer_phone):
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.order_id = order_id
        self.order_date = order_date
        self.order_price = order_price
        self.order_destination = order_destination

    def order_details(self):
        print('Order Details')
        print('-'*55)
        print(f'Order Id: {self.order_id}')
        print(f'Order Date: {self.order_date}')
        print(f'Order Address: {self.order_price}')
        print(f'Order Destination: {self.order_destination}')
```

```python
# customer
# Grandparent class
class Customer:
    def __init__(self, customer_id, customer_name, customer_address, customer_phone):
      self.customer_id = customer_id
      self.customer_name = customer_name
      self.customer_address = customer_address
      self.customer_phone = customer_phone

    def customer_details(self):
        print('Customer Details')
        print('-'*55)
        print(f'Customer Id: {self.customer_id}')
        print(f'Customer Name: {self.customer_name}')
        print(f'Customer Address: {self.customer_address}')
        print(f'Customer Phone: {self.customer_phone}')
```

```python
# product
#parent class
class Product(Customer):
    def __init__(self, product_id, product_name, product_desc, product_quantity, product_price, customer_id, customer_name, customer_address, customer_phone):
        super().__init__(customer_id, customer_name, customer_address, customer_phone)
        self.product_id = product_id
        self.product_name = product_name
        self.product_desc = product_desc
        self.product_quantity = product_quantity
        self.product_price = product_price

    def product_details(self):
        print('Product Details')
        print('-'*55)
        print(f'Product Id: {self.product_id}')
        print(f'Product Name: {self.product_name}')
        print(f'Product Description: {self.product_desc}')
        print(f'Product Quantity: {self.product_quantity}')
        print(f'Product Price: {self.product_price}')
```

```python
# Order
# child class
class Order(Product):
    def __init__(self, order_id, order_date, order_price, order_destination, product_id, product_name, product_desc, product_quantity, product_price, customer_id, customer_name, customer_address, customer_phone):
        super().__init__( product_id, product_name, product_desc, product_quantity, product_price, customer_id, customer_name, customer_address, customer_phone)
        self.order_id = order_id
        self.order_date = order_date
        self.order_price = order_price
        self.order_destination = order_destination

    def order_details(self):
        print('Order Details')
        print('-'*55)
        print(f'Order Id: {self.order_id}')
        print(f'Order Date: {self.order_date}')
        print(f'Order Address: {self.order_price}')
        print(f'Order Destination: {self.order_destination}')
```

# Encapsulation

```python
'''
- Data Hididng

Access Modifiers
--------------------
-Public(self.varName)
-Private(self.__varName)
-Protected(self._varName)
'''
```

```python
# School and Student
class School:
    def __init__(self,school_address, school_contact, school_branch):
        self.__school_name = 'ABC School'
        self.__school_address = school_address
        self.__school_contact = school_contact
        self.__school_branch = school_branch

    def __school_details(self): #making the function private as well
        print(f'School Name: {self.__school_name}')
        print(f'School Address: {self.__school_address}')
        print(f'School Contact: {self.__school_contact}')
        print(f'School Branch: {self.__school_branch}')

    def details(self):
        self.__school_details()
```

```python
class Student(School):
    def __init__(self, student_id, student_name, student_contact, student_class, student_section, school_address, school_contact, school_branch ):
      super().__init__(school_address, school_contact, school_branch)
      self.__student_id = student_id
      self.__student_name = student_name
      self.__student_contact = student_contact
      self.__student_class = student_class
      self.__student_section = student_section

    def student_details(self):
        print(f'Student Id: {self.__student_id}')
        print(f'Student Name: {self.__student_name}')
        print(f'Student Contact: {self.__student_contact}')
        print(f'Student Class: {self.__student_class}')
        print(f'Student Section: {self.__student_section}')
```

```python
from datetime import datetime
class Messenger:
    def send_message(self, send_from, send_to, message):
        print('Messenger')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')
        print(f'Sent: {datetime.now()}') 

class Viber:
    def send_message(self, send_from, send_to, message):
        print('Viber')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')
        print(f'Sent: {datetime.now()}') 

class WhatsApp:
    def send_message(self, send_from, send_to, message):
        print('WhatsApp')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')

class Slack:
    def send_message(self, send_from, send_to, message):
        print('Slack')
        print(f'From: {send_from}')
        print(f'To: {send_to}')
        print(f'Message: {message}')
```

```python
def message(obj, send_from, send_to, message):
    obj.send_message(send_from, send_to, message)
```

## Abstraction

```python
# the data are hidden from the user itself, but if the user wants to view the data, the user can view the data
'''
Abstraction -> Interface

- If any class is made abstract class, at least one function of it must be abstract function.
- Abstract Function -> only -> pass keyword <- , there is no return 
- Other normal class implements abstract class.
- If any abstract class has only function then it is also called as interface class.
- Interface Class -> A class without body.
'''
```

```python
# Abstract Base Class
from abc import ABC, abstractmethod # to make function, we need (abstractmethod), to make the object we need the abstract base class(abc import ABC)
```

```python
#interface class
class Bank(ABC):
    @abstractmethod
    def userDetails(self, amount, account_number):
        pass

    @abstractmethod
    def connection(self):
        pass
```

```python
#Normal class implements abstract class, not inherit them, implement
class MobileBanking(Bank):
    def mobileBankingConnection(self):
        return f'Connected to mobile banking server..'

    def userDetails(self, amount, account_number):
        return f'Rs. {amount} remaining in A/C no. {account_number}.'

    def connection(self):
        return f'Connected to Banking Database Server...'
```
