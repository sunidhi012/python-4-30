# Data Structure

```python
''' 
Data Structure
-----------------------
1. List -> Application Level -> update,delete,add    []
2. Tuple -> If want to change then go to the prog ()
3. Dictionary -> Mostly used for Data and application -> JSON (Javascript Object Notation)  {}
4. Sets -> Unique Records        {}

a. Mutable Data Structure -> Data Structure which can be changed
     -> List, Dictionary, Sets
b. Immutable Data Structure -> Data Structure which data cannot be changed
     -> Tuple
'''
```

## Dictionary

```python
'''
- To define dictionary data we use {}.
- Dictionary does not contain any indexing. It contains key and value pair.
'''
```

```python
product_details = {
    'product_name' : 'Samsung Mobile',
    'price' : 150000,
    'RAM' : '12 GB',
    'storage' : '256 GB',
    'brand' : 'Samsung'
}
product_details
```

```python
product_details_list = {
    'product_name' : ['Samsung Mobile', 'Acer Laptop'],
    'price' : [150000, 130000],
    'RAM' : ['12 GB', '16 GB'],
    'storage' : ['256 GB', '1 TB'],
    'brand' : ['Samsung', 'Acer']
}
product_details_list
```

```python
# Application releted work
# Dictionary - JSON (Javascript Object Notation) Data
# Multiple Dictionary inside list is known as JSON data.
product_json = [
    {
    'product_name' : 'Samsung Mobile',
    'price' : 150000,
    'RAM' : '12 GB',
    'storage' : '256 GB',
    'brand' : 'Samsung'
    },
     {
    'product_name' : 'Redmi Mobile',
    'price' : 50000,
    'RAM' : '12 GB',
    'storage' : '256 GB',
    'brand' : 'Redmi'
    }
    
]
product_json
```

```python
product_details_list
```

```python
## Loop to print dictionary values
'''
.keys() -> only to print keys of dictionary
.values() -> only to print values of dictionary
.items() -> To print both keys and values of dictionary
'''
```

```python
for key in product_details_list.keys():
    print(key)
```

```python
for val in product_details_list.values():
    print(val)
```

```python
for key, val in product_details_list.items():
    print(f'{key}: {val}')
```

```python
for key, val in product_details_list.items():
    print(f'\n{key}')
    print('-' * 15)
    for data in val:
        print(data)
```

## Part 3: Application Level Logic

### Lab Question 5: The "Price Is Right" (Conditional Logic) 

```python
prices = [100, 250, 75, 400, 120, 310]
prices
```

```python
#1. Write a program that creates two new empty lists: cheap_items and expensive_items.
cheap_items = []
expensive_items = []
cheap_items
expensive_items
```

```python
#2. Loop through the prices list. If a price is greater than 200, append it to expensive_items. Otherwise, append it to cheap_items.
for price in prices:
    if price > 200:
        expensive_items.append(price)
    else:
        cheap_items.append(price)
```

```python
#3. Print both lists and calculate the total count of items in each using the len() function.
print("Cheap Items:", cheap_items)
print("Count:", len(cheap_items))

print("Expensive Items:", expensive_items)
print("Count:", len(expensive_items))
```
