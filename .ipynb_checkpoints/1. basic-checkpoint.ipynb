### Input function

```python
'''
integer -> it does not accept float value.
float -> accepts integer and float value both.
'''
```

```python
#input() -> default string value accepts

a=float(input('enter first number:'))
b=float(input('enter second number:'))
c=a+b
print(c)
```

```python
'''
Variable naming convention
-----------------------------------
1. snake case
-> first_name,last_name, years_of_experience,monthly_total_funded_amount_by_user
2. camel case
-> firstName, lastName, yearsOfExperience, monthlyTotalFundedAmountByUser
'''
```

```python
'''
operators
---------------------------
1.Arithmetic operator
-> (+,-,*,/,%(modulus/remainder),//(float division))

2.Logical operator
-> AND(&), OR(|), NOT(!)

3. Comparision operator
-> =,==,>,>=,<,<=,!=
'''
```

```python
num1 =int(input('enter first number:'))
num2 =int(input('enter second number:'))

add= num1 + num2
sub= num1 - num2
mul= num1 * num2
div= num1 / num2
mod= num1 % num2
fd= num1 // num2

print(f'\naddition of {num1} and {num2}: {add}')
print(f'Sudstraction of {num1} and {num2}: {sub}')
print(f'Multiplication of {num1} and {num2}: {mul}')
print(f'Division of {num1} and {num2}: {div}')
print(f'Reamainder/Modulus of {num1} and {num2}: {mod}')
print(f'Float division of {num1} and {num2}: {fd}') #removes the decimal part and only print sthe integer part
```

```python
#Logical operator and comparision operator gives output in Boolean values(0/1).
#0 is false and 1 is true
num1 > num2
```
