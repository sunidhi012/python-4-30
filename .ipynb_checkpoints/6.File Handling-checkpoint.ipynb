# Text File

```python
# Creating text file
# open(file_name,operation)
file = open('textfile.txt', 'w')
content = file.write('Hello World!!!')
file.close()
```

```python
def textFile():
    file_name = input('ENter your file name:')
    with open(f' {file_name}.txt','w') as file:
      content = input('Enter your content:')
      file.write(content)
```

```python
y_n = input('Do you really want to create a file? (y/n):')
if y_n == 'y':
    textFile()
else:
    print('File creation cancelled')
```

```python
# Read text file
with open('textfile.txt','r') as file:
    data = file.read()
    print(data)
```

```python
with open('textfile.txt','a') as file:
    content = input('Enter your content:')
    file.write(f'\n\t{content}')
```

## CSV File

```python
'''
CSV -> Comma Seperated Value (text file)
It opens in excel.Excel is a binary file.
'''
```

```python
import csv
```

```python
products = [
    ['product_id', 'product_name', 'quantity', 'price'],
    [1, 'Mobile', '3', '$500'],
    [2, 'Bulb', '20', '$100'],
    [1, 'Bottle', '1', '$1.25'],
]
```

```python
#open(file_name, operation, newline)
with open('products.csv','w',newline='') as file:
    writer = csv.writer(file) # create csv file
    writer.writerows(products) #write data in csv file
```

```python
# Read CSV data
with open('products.csv','r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
new_products = [
    [4, 'TV', '3', '$3000'],
    [5, 'PS5', '20', '$650'],
    [6, 'Bag', '1', '$20'],
]
```

```python
with open('products.csv','a', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(new_products)
```

```python
employee_details = [
    {
        'Name' : 'Bob Miles',
        'Designation' : 'Senior',
        'Salary' : 200000,
        'Department' : 'Sales'
    },
    {
         'Name' : 'Bill Morgan',
         'Designation' : 'Intern',
         'Salary' : 10000,
         'Department' : 'Marketing'
    },
    {
         'Name' : 'Dutch Marston',
         'Designation' : 'Junior',
         'Salary' : 50000,
         'Department' : 'IT'
    }
]
        
```

```python
# open(file_name, operation, newline)
with open('employees.csv', 'w', newline = '') as file: 
    columns = ['Name' ,'Designation', 'Salary', 'Department']    #column names
    writer = csv.DictWriter (file, fieldnames= columns)     # create file and initialize column names
    writer.writeheader()       #write column names in file  # always write at top of writerows else the header will come at last
    writer.writerows(employee_details)      #write rows in files
```

```python
## Read data
with open ('employees.csv', 'r') as file:
    reader = csv.DictReader(file)
    for row in reader:
        for idx, val in row.items():
            print(f'{idx} : {val}')
            print('-'*20)
```

```python
new_employee_details = [
    {
        'Name' : 'John Doe',
        'Designation' : 'Mid- level',
        'Salary' : 150000,
        'Department' : 'BI reporting'
    }
]
        
```

```python
with open('employee.csv', 'a', newline ='') as file:
    columns = ['Name', 'Designation', 'Salary', 'Department']
    writer = csv.DictWriter(file, fieldnames = columns)
    writer.writerows(new_employee_details)
```

## JSON Data(Javascript Object Notation)

```python
import json
```

```python
#write -> dump()
#read -> load()
#append -> append() + dump()
```

```python
employee_details = [
    {
        'Name' : 'Bob Miles',
        'Designation' : 'Senior',
        'Salary' : 200000,
        'Department' : 'Sales'
    },
    {
         'Name' : 'Bill Morgan',
         'Designation' : 'Intern',
         'Salary' : 10000,
         'Department' : 'Marketing'
    },
    {
         'Name' : 'Dutch Marston',
         'Designation' : 'Junior',
         'Salary' : 50000,
         'Department' : 'IT'
    },
     {
        'Name' : 'John Doe',
        'Designation' : 'Mid- level',
        'Salary' : 150000,
        'Department' : 'BI reporting'
    }
]
        
```

```python
# dump(data, file, indent=4)
with open('employees.json', 'w') as file:
    json.dump(employee_details, file, indent=4)
```

```python
## Read json data
with open('employees.json', 'r') as file:
    data = json.load(file)
    for row in data:
        for idx, val in row.items():
            print(f'{idx} : {val}')
        print('-' *20)
```

```python
employee_json =  {
        'Name' : 'Charles Miles',
        'Designation' : 'Junior',
        'Salary' : 40000,
        'Department' : 'IT'
    }
        
```

```python
data.append(employee_json)
```

```python
with open('employees.json', 'w') as file:
    json.dump(data, file, indent=4)
```
