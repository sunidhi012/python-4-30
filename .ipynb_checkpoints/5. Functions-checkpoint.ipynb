# Functions

```python
'''
Functions name is always in small letters, never capital.
To reuse the same code everytime
- Function is like a component.
- Write code inside  function once and run it to multiple set of code at once.

syntax
--------------
def -> define
def function_name():
 logic...
 print(...)
 return.....
'''
```

```python
'''
Types of function
------------------------
1. Argumented / Parameterized functions
   a. Default Argument Function
   b. Keyword Argument Function
   c. All Argument Function (*args)
   d. All Keyword Argument Function (**kwargs)
   
2. Non Argumented / Non Parameterized Functions
3. Return Type and Non Return Type Functions
4. Lambda Function
'''
```

```python
# Non-Argumented and non return type function
def greet():
    print('Hello World!!') #in order to get the value of the function we must call the function
    # we can put as many value we want to print
```

```python
# call-back function
greet()        # no argumentes must be passed
```

```python
# Argumented Function
def greetPerson(name):
    print(f'Hello {name}')
```

```python
greetPerson('Charles')   #arguments must be necessarily passed
```

```python
def greetReturn():
    return'Hello World!!' # we can only return one value no multiple if want to return multiple then use comma
```

```python
greetReturn()
```

```python
def greetPersonReturn(name):
    return f'Hello {name}'
```

```python
greetPersonReturn('Bob')
```

```python
print(greetReturn())
```

```python
print(greet())
```

```python
''' 
When to use return type function?
-> If we need to call our function from from one file to another then use return type function.
-> we can use print function to display output.
-> Only one return statement can be defined in one single function.


When to use non return type function?
-> If we need to call our function in same file where we defined it then use non return type function.
-> we cannot use print function to display output. Just nedd to call function.
-> Multiple print statements can be defined in one single function.
'''
```

## All Argument Function (*args)

```python
# *args   #mailnly used for data analysis #it is dynamic and many values can be passed
def personalDetails(name,*args):
    for data in args:
     print(data)
    print(name)
```

```python
personalDetails('Ram', 30, 'Bachelors', 'Lalitpur', 50000)
```

## All Keyword Argument Function

```python
# **kwargs
def kwPersonalDetails(**kwargs):
    print(kwargs)
    for key, val in kwargs.items():
        print(f'{key} :{val}')
```

```python
kwPersonalDetails(name='Ram', age=30, qualification='Bachelors',address= 'Lalitpur',salary= 50000)
```

### Nested Function

```python
# Multiple Independent function/ single independent function called by one single function is nested function.
def add(a,b):
    return f'Addition of {a} and {b} is: {a+b}.'

def sub(a,b):
        return f'Subtraction of {a} and {b} is: {a-b}'

def mul(a,b):
            return f'Multiplication of {a} and {b} is: {a*b}'
def div(a,b):
            return f'Division of {a} and {b} is: {a/b}'  
```

```python
def calculator(a,b):
    return add(a,b), sub(a,b), mul(a,b), div(a,b)
```

```python
calculator(1,2)
```

```python
def calc(choice):
    num1 = float(input('Enter a number:'))
    num2 = float(input('Enter a number:'))
    if choice == 1:
        selected_choice = add(num1, num2)
    elif choice == 2:
        selected_choice = sub(num1, num2)
    elif choice == 3:
        selected_choice = mul(num1, num2)
    elif choice == 4:
        selected_choice = div(num1, num2)
    else:
        print('invalid choice')
        
    return selected_choice
```

```python
 calc(2)
```

## Lambda Function

```python
def odd_even(num):
    if num % 2 == 0:
        print('Even')
    else:
        print('Odd')
odd_even(4)
```

```python
#func_name = lambda var : condition / loops
```

```python
o_e = lambda x : "Even" if x % 2 == 0 else "Odd"
```

```python
o_e(56)
```

```python
prices = [100, 250, 75, 400, 120, 310]
cheap_items = []
expensive_items = []
prices = lambda price: "Expensive" if price >= 200 else "Cheap"
prices(100)
```

```python
prices = [100, 250, 75, 400, 120, 310]
cheap_items = []
expensive_items = []
def get_exp(lst):
    for price in lst:
        if price >= 200:
            expensive_items.append(price)

def get_che(lst):
    for price in lst:
        if price < 200:
            cheap_items.append(price)           

get_exp(prices)
get_che(prices)
cheap_items
expensive_items
```

```python
prices = [100, 250, 75, 400, 120, 310]
cheap_items = []
expensive_items = []
get_expensive = lambda lst: [price for price in lst if price >= 200]
get_cheap = lambda lst : [ price for price in lst if price < 200]
get_expensive (prices)
```

```python
get_cheap(prices)
```
